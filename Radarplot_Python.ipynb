```python
%matplotlib inline

# Funtion
def plot_radar_chart(data1, data2, metric, metric2, 
                     title1, title2, main_title):

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (16, 8), subplot_kw = {"projection": "polar"})

    # Define the color palette
    color_palette = ['#559E4B', '#357ABD', '#edbd08', '#734173']
    
    N1 = len(metric)
    theta1 = np.linspace(0, 2 * np.pi, N1, endpoint = False)
    theta1 = np.concatenate([theta1, [theta1[0]]])

    # title
    ax1.set_title(title1, y = 1.15, fontsize = 20)
    # Direction of the zero angle to the north
    ax1.set_theta_zero_location("N")
    # set counterclockwise
    ax1.set_theta_direction(-1)
    # Radial label positiion 
    ax1.set_rlabel_position(90)
    # Make radial gridlines appear behind other elements
    ax1.spines["polar"].set_zorder(1)
    # color of the radial gridline
    ax1.spines["polar"].set_color("lightgrey")
    
    for idx, (i, row) in enumerate(data1.iterrows()):
        values = row[metric].values.flatten().tolist()
        values = values + [values[0]]

    # Add this to plot radar chart lines and markers
        ax1.plot(theta1, values, linewidth=1.75, 
                linestyle = 'solid', 
                label = row['index'], 
                marker = 'o', 
                markersize = 10, 
                color = color_palette[idx % len(color_palette)])
        
        ax1.fill(theta1, values, alpha=0.50, color=color_palette[idx % len(color_palette)])
        
        # plt.yticks([0, 0.25, 0.5, 0.75, 1], ["0", "0.25", "0.5", "0.75", "1"], color="black", size=12)
        ax1.set_xticks(theta1)
        ax1.set_xticklabels(metric + [metric[0]], color='black', size=12)
        ax1.legend(loc='upper right', bbox_to_anchor=(1.2, 1.2))

    # seconde plot
    N2 = len(metric2)
    
    theta2 = np.linspace(0, 2 * np.pi, N2, endpoint = False)
    theta2 = np.concatenate([theta2, [theta2[0]]])

    # fig, ax = plt.subplots(figsize = (12,12),  
    #                      subplot_kw = {"projection": "polar"})

    # title
    ax2.set_title(title2, y = 1.15, fontsize = 20)

    # Direction of the zero angle to the north
    ax2.set_theta_zero_location("N")

    # set counterclockwise
    ax2.set_theta_direction(-1)

    # Radial label positiion 
    ax2.set_rlabel_position(90)

    # Make radial gridlines appear behind other elements
    ax2.spines["polar"].set_zorder(1)

    # color of the radial gridline
    ax2.spines["polar"].set_color("lightgrey")
    
    for idx, (i, row) in enumerate(data2.iterrows()):
        values = row[metric2].values.flatten().tolist()
        values = values + [values[0]]

    # Add this to plot radar chart lines and markers
        ax2.plot(theta2, values, linewidth=1.75, 
                linestyle = 'solid', 
                label = row['index'], 
                marker = 'o', 
                markersize = 10, 
                color = color_palette[idx % len(color_palette)])
        
        ax2.fill(theta2, values, alpha=0.50, color=color_palette[idx % len(color_palette)])
        
        # plt.yticks([0, 0.25, 0.5, 0.75, 1], ["0", "0.25", "0.5", "0.75", "1"], color="black", size=12)
        ax2.set_xticks(theta2)
        ax2.set_xticklabels(metric2 + [metric2[0]], color='black', size=12)
        ax2.legend(loc='upper right', bbox_to_anchor=(1.2, 1.2))

    if main_title:
        fig.suptitle(main_title, fontsize=24, y=0.98)

    plt.tight_layout()

    return fig
```
